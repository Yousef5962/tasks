```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Weather_Dataset.csv')
df.dropna(inplace= True)
df.drop_duplicates(inplace= True)
df = df[(df['Temperature (C)'] >= -50) & (df['Temperature (C)'] <= 50)]
df = df[df['Humidity'] > -1]
```

```python
plt.figure(figsize=(24,8))
plt.plot(df['Formatted Date'], df['Temperature (C)'],color = 'red', linestyle='-.',)
plt.xlabel('Time')
plt.ylabel('Temperature')
plt.title('Temperature Over Time')
plt.ylim(top=50)
plt.show()
```

```python
plt.figure(figsize=(24, 12))
plt.hist(df['Temperature (C)'], edgecolor='k')
plt.xlabel('Temperature')
plt.title('Temperature Distribution')
plt.show()
```

```python
plt.figure(figsize=(16, 8))
plt.scatter(df['Temperature (C)'], df['Humidity'])
plt.xlabel('Temperature')
plt.ylabel('Humidity')
plt.title('Temperature vs. Humidity')
plt.show()
```

```python
ndf = df.select_dtypes(include=[float])
corr = ndf.corr()
sns.heatmap(corr,annot=True,fmt="0.3f")
plt.title('Correlation Heatmap')
plt.show()
```
